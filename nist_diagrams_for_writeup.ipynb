```python
from graphviz import Digraph
dot = Digraph(name='training_flow')
dot.format='png'
L=[('a', 'BUILD audio', 'folder'),
   ('b', 'BUILD transcription', 'folder'),
   ('c', 'Clip to <10s', 'rectangle'),
   ('d', '(audio,gold) pair', 'note'),
   ('e', 'augmentation', 'rectangle'),
   ('f', 'pair + 10 augmentations', 'note'),
  ('g', 'trainer', 'rectangle'),
  ('h', 'trained model', 'cylinder')]
for name, label, shape in L:
    dot.node(name,label, shape=shape)
dot.edges(['ac', 'bc', 'cd', 'de', 'ef', 'fg', 'gh'])
dot
```

```python
dot.render()
```
